```python
# Create pandas dataframe
import pandas as pd
sr_df = pd.DataFrame(food_items)
sr_df.columns = ['ndbid', 'name', 'food_group']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
CountVectorizer?
```

```python
countvect = CountVectorizer(stop_words="english", 
                            strip_accents="ascii", 
                            decode_error= 'ignore')
# creates character n-grams only from text inside word boundaries.  
```

```python
vectorized = countvect.fit_transform(sr_df.name)
```
